```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np
```

```python
df_202201 = pd.read_csv(r'202201_Transferencias.csv', sep=';', encoding='cp1252')
df_202202 = pd.read_csv(r'202202_Transferencias.csv', sep=';', encoding='cp1252')
df_202203 = pd.read_csv(r'202203_Transferencias.csv', sep=';', encoding='cp1252')
df_202204 = pd.read_csv(r'202204_Transferencias.csv', sep=';', encoding='cp1252')
df_202205 = pd.read_csv(r'202205_Transferencias.csv', sep=';', encoding='cp1252')
df_202206 = pd.read_csv(r'202206_Transferencias.csv', sep=';', encoding='cp1252')
```

```python
df_202201.info()
```

```python
new_column_names = ['ano_mes', 'tipo_transferencia', 'tipo_favorecido', 'uf', 'cod_municipio_siafi', 'nome_municipio',
                    'codigo_orgao_siafi', 'nome_orgao', 'codigo_unidade_gestora', 'nome_unidade_gestora', 'codigo_funcao',
                    'nome_funcao', 'codigo_subfuncao', 'nome_subfuncao', 'codigo_rograma', 'nome_programa', 'acao',
                    'nome_acao', 'linguagem_cidada', 'codigo_grupo_despesa', 'nome_grupo_despesa',
                    'codigo_modalidade_aplicacao_despesa','nome_modalidade_aplicacao_despesa', 'codigo_elemento_despesa',
                    'nome_elemento_despesa', 'codigo_plano_orcamentario', 'nome_plano_orcamentario', 'codigo_subtitulo',
                    'nome_subtitulo', 'codigo_localizador', 'nome_localizador', 'sigla_localizador',
                    'descricao_complementar_localizador', 'codigo_favorecido', 'nome_favorecido', 'valor_transferido']

df_202201.columns = df_202202.columns = df_202203.columns = df_202204.columns = df_202205.columns = df_202206.columns = new_column_names
```

```python
df = pd.concat([pd.DataFrame(), df_202201[df_202201['nome_municipio'] == 'CUBATAO']])
df = pd.concat([df, df_202202[df_202202['nome_municipio'] == 'CUBATAO']])
df = pd.concat([df, df_202203[df_202203['nome_municipio'] == 'CUBATAO']])
df = pd.concat([df, df_202204[df_202204['nome_municipio'] == 'CUBATAO']])
df = pd.concat([df, df_202205[df_202205['nome_municipio'] == 'CUBATAO']])
df = pd.concat([df, df_202206[df_202206['nome_municipio'] == 'CUBATAO']])
```

```python
df = df.drop(columns=['uf', 'nome_municipio', 'cod_municipio_siafi', 'codigo_orgao_siafi', 'codigo_unidade_gestora',
                      'codigo_funcao', 'codigo_subfuncao', 'codigo_rograma', 'codigo_grupo_despesa',
                      'codigo_modalidade_aplicacao_despesa', 'codigo_elemento_despesa', 'codigo_plano_orcamentario',
                      'codigo_subtitulo', 'codigo_localizador', 'codigo_favorecido', 'acao'])
```

```python
df.info()
```

```python
df['valor_transferido'] = df['valor_transferido'].str.replace(',','.')
df['valor_transferido'] = df['valor_transferido'].astype('float')
df['valor_transferido'].info()
```

```python
df['ano_mes'] = df['ano_mes'].replace(202201, 'Janeiro').replace(202202, 'Fevereiro').replace(202203, 'Março').replace(202204, 'Abril').replace(202205, 'Maio').replace(202206, 'Junho').replace(202202, 'Fevereiro')
```

## Análise Multivariada

```python
# nome_funcao x valor_transferido

def etiqueta_numeral(valor):
    if valor > 1000000000000:
        return f'R$ {round(valor/1000000000000)} Tri'
    elif valor > 1000000000:
        return f'R$ {round(valor/1000000000)} Bi'
    elif valor > 1000000:
        return f'R$ {round(valor/1000000)} Mi'
    elif valor > 1000:
        return f'R$ {round(valor/1000)} Mil'
    else:
        return f'R$ {round(valor)}'

new_df = df[['nome_funcao', 'valor_transferido']]

data = {
    'nome_funcao': [],
    'valor_transferido': []
}

for nome_funcao in new_df['nome_funcao'].unique():
    temp_df = new_df[new_df['nome_funcao'] == nome_funcao]
    data['nome_funcao'].append(str(temp_df['nome_funcao'].iloc[0]))
    data['valor_transferido'].append(temp_df['valor_transferido'].sum())

new_df = pd.DataFrame(data)
new_df = new_df.sort_values('valor_transferido', ascending=False)

x = new_df['nome_funcao']
y = new_df['valor_transferido']

fig, ax = plt.subplots(figsize=(10,5), dpi=80)
plt.bar(x, y, color='green')

plt.title('Valor Transferido Por Função')
plt.xlabel('Função')
plt.ylabel('Valor Transferido (R$)')

ax.ticklabel_format(style='plain', axis='y')
escala_y = plt.gca().get_yticks()
ax.yaxis.set_major_locator(mticker.FixedLocator(escala_y))
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in escala_y])

rects = ax.patches
labels = [etiqueta_numeral(valor) for valor in new_df['valor_transferido']]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + (height*0.005), label, ha="center", va="bottom")

plt.show()
```

```python
# UNDER ANALYSIS

new_df = df[['ano_mes', 'nome_funcao', 'valor_transferido']]
data = {
    'ano_mes': [],
    'nome_funcao': [],
    'valor_transferido': []
}

for ano_mes in new_df['ano_mes'].unique():
    temp_df1 = new_df[new_df['ano_mes'] == ano_mes]
    for nome_funcao in temp_df1['nome_funcao'].unique():
        temp_df2 = temp_df1[temp_df1['nome_funcao'] == nome_funcao]
        data['ano_mes'].append(temp_df2['ano_mes'].iloc[0])
        data['nome_funcao'].append(temp_df2['nome_funcao'].iloc[0])
        data['valor_transferido'].append(temp_df2['valor_transferido'].sum())

new_df = pd.DataFrame(data)

x = new_df['ano_mes'].unique()
y = []

counter = 0
for nome_funcao in new_df['nome_funcao'].unique():
    y.append([])
    for mes in x:
        if mes in list(new_df[new_df['nome_funcao'] == nome_funcao]['ano_mes']):
            y[counter].append(new_df[(new_df['ano_mes'] == mes) & (new_df['nome_funcao'] == nome_funcao)]['valor_transferido'].iloc[0])
        else:
            y[counter].append(0)
    counter += 1

fig, ax = plt.subplots(figsize=(10,6), dpi=80)

plt.title('Valor Transferido Por Função')
plt.xlabel('Meses')
plt.ylabel('Valor Transferido (R$)')
    
line1 = plt.plot(x, y[0], marker='o')
line2 = plt.plot(x, y[1], marker='o')
line3 = plt.plot(x, y[2], marker='o')
line4 = plt.plot(x, y[3], marker='o')
line5 = plt.plot(x, y[4], marker='o')

ax.ticklabel_format(style='plain', axis='y')
escala_y = plt.gca().get_yticks()
ax.yaxis.set_major_locator(mticker.FixedLocator(escala_y))
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in escala_y])

plt.legend((line1[0], line2[0], line3[0], line4[0], line5[0]), new_df['nome_funcao'].unique())

ax.set_axisbelow(True)
ax.grid(color='lightgray', linestyle='dashed')

#ax2.set_xlabel('Strain (in/in)')
#ax2.set_ylabel('Stress (ksi)')
#ax2.set_title('Inset of Elastic Region')
#ax2.set_xlim([0,0.008])
#ax2.set_ylim([0,100])

plt.show()
```
