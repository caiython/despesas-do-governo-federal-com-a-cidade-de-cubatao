```python
df.info()
```

```python
df['valor_transferido'] = df['valor_transferido'].str.replace(',','.')
df['valor_transferido'] = df['valor_transferido'].astype('float')
df['valor_transferido'].info()
```

# Análise Univariada

```python
meses_dict = {
    'Janeiro': 1,
    'Fevereiro': 2,
    'Março': 3,
    'Abril': 4,
    'Maio': 5,
    'Junho': 6
}

new_df = df['ano_mes'].value_counts().rename_axis('ano_mes').reset_index(name='contagem').sort_values('ano_mes', key = lambda x: x.apply (lambda x : meses_dict[x]))

x = new_df['ano_mes']
y = new_df['contagem']

fig, ax = plt.subplots(figsize=(10,5), dpi=80)
plt.bar(x, y, color='turquoise')

plt.title('Número de Lançamentos por Mês')
plt.xlabel('Mês')
plt.ylabel('Número de Lançamentos')

ax.bar_label(ax.containers[0], label_type='edge')

plt.show()
```

# Análise Multivariada

```python
# nome_funcao x valor_transferido

def etiqueta_numeral(valor):
    if valor > 1000000000000:
        return f'{round(valor/1000000000000)} Tri'
    elif valor > 1000000000:
        return f'{round(valor/1000000000)} Bi'
    elif valor > 1000000:
        return f'{round(valor/1000000)} Mi'
    elif valor > 1000:
        return f'{round(valor/1000)} Mil'
    else:
        return f'{round(valor)}'

new_df = df[['nome_funcao', 'valor_transferido']]

data = {
    'nome_funcao': [],
    'valor_transferido': []
}

for nome_funcao in new_df['nome_funcao'].unique():
    temp_df = new_df[new_df['nome_funcao'] == nome_funcao]
    data['nome_funcao'].append(str(temp_df['nome_funcao'].iloc[0]))
    data['valor_transferido'].append(temp_df['valor_transferido'].sum())

new_df = pd.DataFrame(data)
new_df = new_df.sort_values('valor_transferido', ascending=False)

x = new_df['nome_funcao']
y = new_df['valor_transferido']

fig, ax = plt.subplots(figsize=(10,5), dpi=80)
plt.bar(x, y, color='green')

plt.title('Valor Transferido Por Função')
plt.xlabel('Função')
plt.ylabel('Valor Transferido')

rects = ax.patches
labels = [etiqueta_numeral(valor) for valor in new_df['valor_transferido']]

for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label, ha="center", va="bottom")

plt.show()
```
